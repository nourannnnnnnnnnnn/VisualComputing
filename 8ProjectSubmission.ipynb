```python
import cv2
import numpy as np
import face_recognition
import os
from datetime import datetime
```

```python
path = 'Downloads/AttendanceImages/'
images = [] #empty list to carry list of images 
classNames = []#className for each image as each image represent class
myList = os.listdir(path)#passing images path to var myList
print(myList) 
```

```python
for cl in myList: #Looping over images list
    curImg = cv2.imread(f'{path}/{cl}')#enter path and get images inside it
    images.append(curImg)#
    classNames.append(os.path.splitext(cl)[0])
print(classNames)
```

```python
def findEncodings(images):
    encodeList = [] 
    for img in images:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)#convert img from bgr to rgb
        encode = face_recognition.face_encodings(img)[0]#recognition for face_encodinged images
        encodeList.append(encode)
    return encodeList
```

```python
def markAttendance(name):
    with open('Downloads/a.csv','r+') as f:
        myDataList = f.readlines()
        nameList = []
        for line in myDataList:
            entry = line.split(',')
            nameList.append(entry[0])
        if name not in nameList:
            now = datetime.now()
            dtString = now.strftime('%H:%M:%S')
            f.writelines(f'\n{name},{dtString}')
        

encodeListKnown = findEncodings(images)
print('Encoding Complete')
```

```python
cap = cv2.VideoCapture(0)
frame_resizing = 0.25

while True:
    success, img = cap.read()
    imgS = cv2.resize(img, (0, 0), fx=frame_resizing, fy=frame_resizing)
    imgS = cv2.cvtColor(imgS, cv2.COLOR_BGR2RGB)
 
    facesCurFrame = face_recognition.face_locations(imgS)
    encodesCurFrame = face_recognition.face_encodings(imgS,facesCurFrame)
 
    for encodeFace,faceLoc in zip(encodesCurFrame,facesCurFrame):
        matches = face_recognition.compare_faces(encodeListKnown,encodeFace) #comparing stored images with images coming from real time
        print('matches',matches)
        faceDis = face_recognition.face_distance(encodeListKnown,encodeFace) #Get Distance 
        print(faceDis)
        matchIndex = np.argmin(faceDis)#Get index of the smallest distance image
        if matches[matchIndex]:
            name = classNames[matchIndex].upper()
            print(name)
            
            faceLoc = np.array(faceLoc)
            faceLoc = faceLoc / 0.25
            faceLoc=faceLoc.astype(int)
          
            y1, x2, y2, x1 = faceLoc[0], faceLoc[1], faceLoc[2], faceLoc[3] #top right ,botom left
         
            cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),2)#draw rectangle of the name
            cv2.rectangle(img,(x1,y2-35),(x2,y2),(0,255,0),cv2.FILLED)#draw rectangle of the image
            cv2.putText(img,name,(x1+6,y2-6),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),2)
            markAttendance(name)#call function called "markAttendance "
 
    cv2.imshow('Webcam',img)
    key = cv2.waitKey(1)
    if key == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
